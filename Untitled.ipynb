```python
from pandas.io.json import json_normalize
import pandas as pd

pd.options.display.max_columns = 500

orders_df = pd.DataFrame(orders)
```

```python
def extract_and_flatten(df, col_name_to_extract, original_id_col, original_id_col_rename, extracted_df_key):

    extracted = df[col_name_to_extract].apply(pd.Series)
    extracted = extracted.merge(orders_df[[original_id_col]], left_index=True, right_index=True)
    extracted_melt = extracted.melt(id_vars=original_id_col).drop('variable', axis=1)
    extracted_melt = extracted_melt.rename({original_id_col:original_id_col_rename}, axis=1)
    flat = json_normalize(json.loads(extracted_melt.to_json(orient='records')))
    flat.columns = [col.replace('value.', '') for col in flat.columns]
    
    return flat.loc[~pd.isnull(flat[extracted_df_key])]
    
```

```python
%%timeit
line_items = extract_and_flatten(
    df=orders_df, 
    col_name_to_extract='line_items',
    original_id_col='id',
    original_id_col_rename='check_id',
    extracted_df_key='id'
)
```

```python
# %%timeit
line_items = []
applied_taxes = []
modifications = []

for order in orders:
        
    for line_item, tax in zip(order['line_items'], order['applied_taxes']):
        line_item['order_id'] = order['id']
        tax['order_id'] = order['id']
        
        line_items.append(line_item)
        applied_taxes.append(tax)
        for mod in line_item['modifications']:
            mod['line_item_id'] = line_item['id']
            modifications.append(mod)
            
```
